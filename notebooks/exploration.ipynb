```python
from datasets import load_dataset
import matplotlib.pyplot as plt
import numpy as np

ds = load_dataset("blanchon/EuroSAT_RGB")
print(ds)
```

```python
CLASS_NAMES = [
    "AnnualCrop", "Forest", "HerbaceousVegetation", "Highway", "Industrial",
    "Pasture", "PermanentCrop", "Residential", "River", "SeaLake"
]

fig, axes = plt.subplots(2, 5, figsize=(15, 6))
for i, ax in enumerate(axes.flat):
    sample = ds["train"][i]
    ax.imshow(np.array(sample["image"]))
    ax.set_title(CLASS_NAMES[sample["label"]])
    ax.axis("off")
plt.tight_layout()
plt.show()
```

```python
from torchvision import transforms
import torch

transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])

sample_image = ds["train"][0]["image"].convert("RGB")
tensor = transform(sample_image)
print(f"Input shape: {tensor.shape} | dtype: {tensor.dtype} | min: {tensor.min():.2f} | max: {tensor.max():.2f}")
```

```python
import timm

model = timm.create_model("efficientnet_b0", pretrained=True, num_classes=10)
model.eval()

with torch.no_grad():
    logits = model(tensor.unsqueeze(0))

print(f"Output shape: {logits.shape}")
print(f"Predicted class: {CLASS_NAMES[logits.argmax().item()]}")
```
